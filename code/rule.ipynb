```python
###生成测试集的数据
def generate(df_city,city):
    arr_time=pd.date_range(start = '11/16/2018',end='2/25/2019',freq = 'h').astype(np.str).values[:-1]
    #print(arr_time.shape)
    arr_time=arr_time.reshape([arr_time.shape[0],1])
    list_city=[]
    list_value=[]
    for i in range(arr_time.shape[0]):
        list_city.append(city)
        list_value.append(1.0)
    arr_city=np.array(list_city)
    arr_city=arr_city.reshape([arr_city.shape[0],1])
    #arr_value=np.empty_like(arr_city)
    arr_value=np.array(list_value)
    arr_value=arr_value.reshape([arr_value.shape[0],1])
    arr_extend=np.concatenate((arr_time,arr_city,arr_value),axis=1)
    arr_city=df_city.values
    #print(arr_city.shape)
    arr_return=np.concatenate((arr_city,arr_extend),axis=0)
    list_connect=[]
    for i in range(arr_return.shape[0]):
        list_temp=[]
        year=datetime.strptime(arr_return[i][0],"%Y-%m-%d %H:%M:%S").strftime("%Y")
        month=datetime.strptime(arr_return[i][0],"%Y-%m-%d %H:%M:%S").strftime("%m")
        day=datetime.strptime(arr_return[i][0],"%Y-%m-%d %H:%M:%S").strftime("%d")
        weekday=datetime.strptime(arr_return[i][0],"%Y-%m-%d %H:%M:%S").strftime("%w")
        hour=datetime.strptime(arr_return[i][0],"%Y-%m-%d %H:%M:%S").strftime("%H")
        list_temp.append(year)
        list_temp.append(str(int(month)))
        list_temp.append(str(int(day)))
        list_temp.append(weekday)
        list_temp.append(str(int(hour)))
        list_connect.append(list_temp)
    arr_connect=np.array(list_connect)
    #print(arr_connect.shape)
    arr_index=np.array(range(arr_return.shape[0])).reshape([arr_return.shape[0],1])
    arr_return=np.concatenate((arr_index,arr_return,arr_connect),axis=1)    
    df_return=pd.DataFrame(arr_return,columns=["索引","时间","地市","流量","年份","月份","几号","周几","几点"])
    for i in range(16392,16464):
        #print(df_return.iat[i,1])
        df_return.iat[i,3]=(df_return.iat[i-24,3]+df_return.iat[i-24*7,3])/2    
    return df_return
df_temp=generate(df_data_A,"A")
#print(df_temp.iloc[16392:16466,3])
df_temp
```

```python
##加入规则一特征（滑动窗口选8）
def add_rule1(df_temp):
    print(df_temp.shape)
    arr_list=[]
    #week_list=["1"]
    week_list=[str(i) for i in range(7)]
    for weekday in week_list:
        df_weekday=df_temp.loc[df_temp["周几"]==weekday]
        arr_result=df_weekday.values
        arr_weekday=df_weekday["流量"].values
        arr_weekday_list=[]
        for i in range(arr_weekday.shape[0]//24):
            arr_weekday_list.append(arr_weekday[i*24:(i+1)*24].reshape([1,24]))
        arr_weekday_get=arr_weekday_list[0]
        for i in range(1,len(arr_weekday_list)):
            arr_weekday_get=np.concatenate((arr_weekday_get,arr_weekday_list[i]),axis=0)
        arr_weekday_get=arr_weekday_get.astype(np.float64)
        arr_sum=np.sum(arr_weekday_get,axis=1).reshape([arr_weekday_get.shape[0],1])
        arr_weekday_get=arr_weekday_get/arr_sum
        arr_pre_list=[]
        arr_pre_list.append(((arr_weekday_get[0,:])*arr_sum[0,0]).reshape([24,1]))
        arr_pre_list.append(((arr_weekday_get[0,:])*arr_sum[0,0]).reshape([24,1]))
        arr_pre_list.append(((arr_weekday_get[1,:])*arr_sum[1,0]).reshape([24,1]))
        arr_pre_list.append(((arr_weekday_get[2,:])*arr_sum[2,0]).reshape([24,1]))
        for i in range(4,8):
            arr_temp=arr_weekday_get[0:i,:]
            arr_med=np.median(arr_temp,axis=0).reshape([24,1])
            arr_pre_list.append(arr_med*arr_sum[i-1,0])
        for i in range(8,98):
            arr_temp=arr_weekday_get[i-8:i,:]
            arr_med=np.median(arr_temp,axis=0).reshape([24,1])
            arr_pre_list.append(arr_med*arr_sum[i-1,0])
        for i in range(98,arr_weekday_get.shape[0]):
            arr_temp=arr_weekday_get[i-8:i,:]            
            arr_med=np.median(arr_temp,axis=0)
            arr_sum[i]=np.mean(arr_sum[i-8:i])
            arr_weekday_get[i,:]=arr_med
            arr_med=arr_med.reshape([24,1])
            arr_pre_list.append(arr_med*arr_sum[i-1,0])
        arr_pre=arr_pre_list[0]
        for i in range(1,len(arr_pre_list)):
            arr_pre=np.concatenate((arr_pre,arr_pre_list[i]),axis=0)
        arr_result=np.concatenate((arr_result,arr_pre),axis=1)
        arr_list.append(arr_result)
    arr_rule1=arr_list[0]
    for i in range(1,len(arr_list)):
        arr_rule1=np.concatenate((arr_rule1,arr_list[i]),axis=0)
    df_rule1=pd.DataFrame(arr_rule1,columns=["索引","时间","地市","流量","年份","月份","几号","周几","几点","规则一(8)"])
    df_rule1=df_rule1.sort_values("索引")
    return df_rule1
df_rule1=add_rule1(df_temp)
df_rule1
```
